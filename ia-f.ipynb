```python
dataset_home = './'
subdirs = ['train_set/', 'val_set/']

for subdir in subdirs:
    labeldirs = ['feliz/', 'triste/', 'raiva/']
    
    for labeldir in labeldirs:
        newdir = dataset_home + subdir + labeldir
        os.makedirs(newdir, exist_ok=True)
```

```python
from sklearn.model_selection import train_test_split

where_is_data = 'train'
all_data = os.listdir(where_is_data)
train_data, val_data = train_test_split(all_data, test_size=0.2)
```

```python
for dst_data, dst_set_name in zip([train_data, val_data], ['train_set/', 'val_set/']):
    for f_name in dst_data:
        if f_name.startswith('feliz'):
            src = 'train/' + f_name
            dst = dst_set_name + 'feliz/' + f_name
            os.rename(src, dst)

        elif f_name.startswith('triste'):
            src = 'train/' + f_name
            dst = dst_set_name + 'triste/' + f_name
            os.rename(src, dst)
        
        elif f_name.startswith('raiva'):
            src = 'train/' + f_name
            dst = dst_set_name + 'raiva/' + f_name
            os.rename(src, dst)
     
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255)
validation_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    'train_set',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

validation_generator = validation_datagen.flow_from_directory(
    'val_set',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)
```

```python
from tensorflow.keras.applications import InceptionV3

model_inception = InceptionV3(
    include_top=False,
    weights="imagenet",
    input_tensor=None,
    input_shape=(224, 224, 3),
    pooling=None,
    classifier_activation="softmax",
)

for layer in model_inception.layers:
    layer.trainable = False
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
     
```

```python
transferred_model= Sequential([
    model_inception,
    Flatten(),
    Dense(512, activation='relu'),
    Dense(3, activation='softmax')
])

transferred_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
     
```

```python
history = transferred_model.fit(
    train_generator,
    epochs=5,
    validation_data=validation_generator
)
     
```

```python
loss, accuracy = transferred_model.evaluate(validation_generator)
print("Inception Model Accuracy: ", accuracy)
```

```python

import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image

def load_and_preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.
    return img_array, img

def display_prediction(img_path, model):
    img_array, img = load_and_preprocess_image(img_path)
    prediction = model.predict(img_array)

    plt.imshow(img)
    plt.axis('off')

    predicted_class = np.argmax(prediction)

    if predicted_class == 0:
        accuracy = prediction[0][predicted_class] * 100
        plt.title("Prediction: feliz {:.2f}%".format(accuracy))
    elif predicted_class == 1:
        accuracy = prediction[0][predicted_class] * 100
        plt.title("Prediction: triste {:.2f}%".format(accuracy))
    else:
        accuracy = prediction[0][predicted_class] * 100
        plt.title("Prediction: raiva {:.2f}%".format(accuracy))


    plt.show()
```
